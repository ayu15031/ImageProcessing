```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("smooth.tif")
# original image
plt.imshow(img)
```

```python
kernel = np.array([[1,1,1],[1,-8,1],[1,1,1]])
dst = cv2.filter2D(img,-1,kernel)
# Laplase filter at all angles
plt.imshow(dst)
kernel
```

```python
# negative 
kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
```

```python
# Laplase filter at 90 degree
kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]])
dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
```

```python
# negative
kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
```

```python
# horizontal laplase filter
kernel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
```

```python
# negative
kernel = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
```

```python
# Vertical Laplase filter
kernel = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
```

```python
# negative
kernel = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
```
