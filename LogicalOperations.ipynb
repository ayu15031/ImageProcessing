```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("a21.png")

# Converting to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# showing the image
plt.imshow(img_gray, cmap='gray')
```

```python
# Reading second image
img2 = cv2.imread("a22.png")

# Converting to grayscale
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Showing image
plt.imshow(img2_gray, cmap='gray')
```

```python
# getting height and width of one image
h, w = img_gray.shape

# resizing 2 image equal to 1 image
print(cv2.resize(img2_gray,(w,h)).shape)

# not logical operation of first image
plt.imshow(cv2.bitwise_not(img_gray), cmap='gray', vmin=0, vmax=255)
```

```python
# not logical operation of second image
plt.imshow(cv2.bitwise_not(cv2.resize(img2_gray,(w,h))), cmap='gray', vmin=0, vmax=255)
```

```python
# and logical operation of 1 and resized 2 image
plt.imshow(cv2.bitwise_and(img_gray,cv2.resize(img2_gray,(w,h))), cmap='gray', vmin=0, vmax=255)
```

```python
# or logical operation of 1 and resized 2 image
plt.imshow(cv2.bitwise_or(img_gray,cv2.resize(img2_gray,(w,h))), cmap='gray', vmin=0, vmax=255)
```
