```python
import cv2
import matplotlib.pyplot as plt

# Reading image
img = cv2.imread("img1.jpg")

# Converting to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# showing the image
plt.imshow(img_gray, cmap='gray')
```

```python
# Reading secnd image
img2 = cv2.imread("download.jpg")

# Converting to grayscale
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Showing image
plt.imshow(img2_gray, cmap='gray')
```

```python
# getting height and width of one image
h, w = img_gray.shape

# Padding in 2 image to make it equal to first one 
img_reshape= cv2.copyMakeBorder(img2_gray,405,405,184,185,cv2.BORDER_CONSTANT,value=[0])

# showing addition of two images using padding
plt.imshow(cv2.add(img_gray,img_reshape), cmap='gray', vmin=0, vmax=255)
```

```python
# resizing 2 image equal to 1 image
print(cv2.resize(img2_gray,(w,h)).shape)

# showing addition of 1 and resized 2 image
plt.imshow(cv2.add(img_gray,cv2.resize(img2_gray,(w,h))), cmap='gray', vmin=0, vmax=255)
```

```python
# subtracting the images
plt.imshow(cv2.subtract(img_gray,cv2.resize(img2_gray,(w,h))), cmap='gray', vmin=0, vmax=255)
```

```python
# multiplying the images
plt.imshow(cv2.multiply(img_gray,cv2.resize(img2_gray,(w,h))), cmap='gray', vmin=0, vmax=255)
```
