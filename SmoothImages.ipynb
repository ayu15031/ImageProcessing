```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("smooth.tif")

# Smoothening using kernel of ones of different sizes

kernel = np.ones((3,3))/9

dst = cv2.filter2D(img,-1,kernel)

kernel = np.ones((5,5))/25

dst2 = cv2.filter2D(img,-1,kernel)

kernel = np.ones((9,9))/81

dst3 = cv2.filter2D(img,-1,kernel)

kernel = np.ones((15,15))/225

dst4 = cv2.filter2D(img,-1,kernel)

kernel = np.ones((35,35))/1225

dst5 = cv2.filter2D(img,-1,kernel)

plt.figure(figsize=(10,10))
plt.subplot(321), plt.imshow(img, cmap="gray"), plt.title("Original Image")
plt.subplot(322), plt.imshow(dst, cmap="gray"), plt.title("3x3 kernel")
plt.subplot(323), plt.imshow(dst2, cmap="gray"), plt.title("5x5 kernel")
plt.subplot(324), plt.imshow(dst3, cmap="gray"), plt.title("9x9 kernel")
plt.subplot(325), plt.imshow(dst4, cmap="gray"), plt.title("15x15 kernel")
plt.subplot(326), plt.imshow(dst5, cmap="gray"), plt.title("35x35 kernel")
```

```python
# Smoothening with a different kernel

kernel = np.array([[1,2,1],[2,4,2],[1,2,1]])/16

dst = cv2.filter2D(img,-1,kernel)



plt.figure(figsize=(6,6))
plt.subplot(121), plt.imshow(img, cmap="gray"), plt.title("Original Image")
plt.subplot(122), plt.imshow(dst, cmap="gray"), plt.title("Smooth image")
```

```python
# Smoothening by non linear operations, taking max in a window size of 3x3
img1 = cv2.copyMakeBorder(img,1,1,1,1,cv2.BORDER_CONSTANT,value=[0])
img2 = img.copy()

for i in range(img1.shape[0]-2):
    for j in range(img1.shape[1]-2):
        img2[i,j] = np.amax(img1[i:i+3,j:j+3])

plt.figure(figsize=(6,6))
plt.subplot(121), plt.imshow(img, cmap="gray"), plt.title("Original Image")
plt.subplot(122), plt.imshow(img2, cmap="gray"), plt.title("Smooth Image")
```

```python
# Smoothening by non linear operations, taking max in a window size of 9x9
for i in range(img1.shape[0]-8):
    for j in range(img1.shape[1]-8):
        img2[i,j] = np.amax(img1[i:i+9,j:j+9])

plt.figure(figsize=(6,6))
plt.subplot(121), plt.imshow(img, cmap="gray"), plt.title("Original Image")
plt.subplot(122), plt.imshow(img2, cmap="gray"), plt.title("Smooth Image")
```

```python
# Smoothening by non linear operations, taking min in a window size of 3x3
for i in range(img1.shape[0]-2):
    for j in range(img1.shape[1]-2):
        img2[i,j] = np.amin(img1[i:i+3,j:j+3])

plt.figure(figsize=(6,6))
plt.subplot(121), plt.imshow(img, cmap="gray"), plt.title("Original Image")
plt.subplot(122), plt.imshow(img2, cmap="gray"), plt.title("Smooth Image")
```
